```python
#Download WordNet resources
import nltk
nltk.download('wordnet')
nltk.download('omw-1.4')
```

```python
#Import WordNet from NLTK
from nltk.corpus import wordnet as wn
```

```python
#Define sample words
word_syn = "happy"
word_ant = "good"
word_hyper = "dog"
```

```python
#Identify synonyms using WordNet
synonyms = set()

for syn in wn.synsets(word_syn):
    for lemma in syn.lemmas():
        synonyms.add(lemma.name())

print("Synonyms of", word_syn, ":")
print(synonyms)
```

```python
#Identify antonyms using WordNet
antonyms = set()

for syn in wn.synsets(word_ant):
    for lemma in syn.lemmas():
        if lemma.antonyms():
            antonyms.add(lemma.antonyms()[0].name())

print("Antonyms of", word_ant, ":")
print(antonyms)
```

```python
#Identify hypernyms using WordNet
hypernyms = set()

for syn in wn.synsets(word_hyper):
    for hyper in syn.hypernyms():
        hypernyms.add(hyper.name())

print("Hypernyms of", word_hyper, ":")
print(hypernyms)
```

```python
#Identify semantic relationships in text
text = "The dog is very happy and good"
words = text.lower().split()

for word in words:
    synsets = wn.synsets(word)
    if synsets:
        print(f"\nWord: {word}")
        print(" Synonyms :", [lemma.name() for lemma in synsets[0].lemmas()])
        print(" Hypernyms:", [h.name() for h in synsets[0].hypernyms()])
```
