```python
#Import required libraries and download WordNet resources
import nltk

nltk.download('punkt')
nltk.download('wordnet')
nltk.download('omw-1.4')

from nltk.wsd import lesk
from nltk.tokenize import word_tokenize
from nltk.corpus import wordnet as wn
```

```python
#Define ambiguous sentences
sentence1 = "I went to the bank to deposit money."
sentence2 = "The river bank was full of sand."
```

```python
import nltk
nltk.download('punkt_tab', quiet=True)

#Apply Lesk Algorithm for Word Sense Disambiguation

# Tokenize sentences
tokens1 = word_tokenize(sentence1)
tokens2 = word_tokenize(sentence2)

# Disambiguate the word 'bank'
sense1 = lesk(tokens1, 'bank')
sense2 = lesk(tokens2, 'bank')

print("Sentence 1:", sentence1)
print("Predicted Sense:", sense1)
print("Definition:", sense1.definition())
print("-" * 50)

print("Sentence 2:", sentence2)
print("Predicted Sense:", sense2)
print("Definition:", sense2.definition())
```

```python
#Display all possible senses of the word 'bank'

all_senses = wn.synsets('bank')

print("All possible senses of 'bank':\n")
for sense in all_senses:
    print(sense, ":", sense.definition())
```
