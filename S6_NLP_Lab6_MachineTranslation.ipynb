```python
#Import required modules
from transformers import MarianMTModel, MarianTokenizer
```

```python
#Load pre-trained English to Hindi translation model
model_name = "Helsinki-NLP/opus-mt-en-hi"

tokenizer = MarianTokenizer.from_pretrained(model_name)
model = MarianMTModel.from_pretrained(model_name)
```

```python
#Define English public information text
english_text = [
    "The government has announced new education policies.",
    "Please maintain cleanliness in public places.",
    "Emergency services are available 24 hours."
]
```

```python
#Translate English text to Hindi
translated = []

for text in english_text:
    inputs = tokenizer(text, return_tensors="pt", padding=True)
    outputs = model.generate(**inputs)
    hindi_text = tokenizer.decode(outputs[0], skip_special_tokens=True)
    translated.append(hindi_text)

for en, hi in zip(english_text, translated):
    print("English :", en)
    print("Hindi   :", hi)
    print("-" * 40)
```

```python
#Save translation output to file
import pandas as pd

df = pd.DataFrame({
    "English_Text": english_text,
    "Hindi_Translation": translated
})

df.to_csv("english_to_hindi_translation.csv", index=False)
print("Translation saved successfully!")
```
