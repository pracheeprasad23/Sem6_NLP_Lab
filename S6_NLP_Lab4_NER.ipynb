```python
#Import necessary libraries
import spacy
from seqeval.metrics import accuracy_score, precision_score, recall_score, f1_score
```

```python
#Load spaCy pre-trained NER model
nlp = spacy.load("en_core_web_sm")
```

```python
#Define real-world text data
texts = [
    "Apple is looking to buy a startup in London for $1 billion",
    "Elon Musk tweeted about Tesla from California",
    "Google announced a new AI lab in India"
]
```

```python
#Perform Named Entity Recognition
predicted_entities = []

for text in texts:
    doc = nlp(text)
    entities = [(ent.text, ent.label_) for ent in doc.ents]
    predicted_entities.append(entities)
    print(f"\nText: {text}")
    print("Entities:", entities)
```

```python
# Title: Define ground truth entities
true_entities = [
    [("Apple", "ORG"), ("London", "GPE"), ("$1 billion", "MONEY")],
    [("Elon Musk", "PERSON"), ("Tesla", "ORG"), ("California", "GPE")],
    [("Google", "ORG"), ("India", "GPE")]
]
```

```python
#Convert entity lists to BIO format
def to_bio(text, entities):
    doc = nlp(text)
    bio_tags = []
    for token in doc:
        tag = "O"
        for ent_text, ent_label in entities:
            if token.text in ent_text.split():
                tag = "B-" + ent_label
        bio_tags.append(tag)
    return bio_tags

y_true = []
y_pred = []

for text, true_ent, pred_ent in zip(texts, true_entities, predicted_entities):
    y_true.append(to_bio(text, true_ent))
    y_pred.append(to_bio(text, pred_ent))
```

```python
#Evaluate NER model using Accuracy, Precision, Recall, and F1-score
accuracy = accuracy_score(y_true, y_pred)
precision = precision_score(y_true, y_pred)
recall = recall_score(y_true, y_pred)
f1 = f1_score(y_true, y_pred)

print("Accuracy :", accuracy)
print("Precision:", precision)
print("Recall   :", recall)
print("F1-score :", f1)
```
